## Importing the dataset

```python
riskData = pd.read_csv('Training Data.csv')
riskData.head()
```

### Data preparation and cleaning and EDA

Converting Risk_Flag into categorical variable as 0 means no risk and 1 means there is risk.
We dropped the Risk_Flag column and replaced it with a Risk_Class column with 0 representing No Risk and 1 representing Risk.

We have done so as Risk_Flag is a categorical data (yes or no) but is given in the dataset in a binary form.

```python
## Converting Risk_Flag from numeric to object
Risk_Class = []
for index, row in riskData.iterrows():
    if row['Risk_Flag'] == 1:
        Risk_Class.append('Risk')
    else:
        Risk_Class.append('No Risk')
riskData['Risk_Class'] = Risk_Class
newRiskData = riskData.drop(columns = 'Risk_Flag')
```

```python
newRiskData.head()
```

We have decided to drop `Id` column as we can use the index of the column instead

```python
newRiskData =  pd.DataFrame(newRiskData.drop(columns = 'Id'))
newRiskData.head()
```
